### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(train,test,test_size=.25,random_state=10)

linear_regression=LinearRegression()

linear_regression.fit(x_train,y_train)

x_train_prediction=linear_regression.predict(x_train)
x_test_prediction=linear_regression.predict(x_test)
```

### Accuracy score for Train Data

```python
from sklearn import metrics

# Mean Absolute error
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_train,x_train_prediction))

# Mean Squared error
print('Mean Squared Error: ', metrics.mean_squared_error(y_train,x_train_prediction))

# Root Mean Squared error
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_train,x_train_prediction)))
```

### Accuracy score for Test Data

```python
# Mean Absolute error
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test,x_test_prediction))

# Mean Squared error
print('Mean Squared Error: ', metrics.mean_squared_error(y_test,x_test_prediction))

# Root Mean Squared error
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test,x_test_prediction)))
```

###  Cross Validation Score

```python
from sklearn.model_selection import cross_val_score

cross=cross_val_score(LinearRegression(),x_train,y_train,cv=10, scoring='neg_mean_squared_error')
print(cross)
print('Cross Validation score: ', np.mean(cross))
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

x_train,x_test,y_train,y_test=train_test_split(train,test,test_size=.25,random_state=10)

dtree=DecisionTreeClassifier()

dtree.fit(x_train,y_train)
print(dtree.feature_importances_)
```

```python
# generating  tree visualization
from sklearn import tree
with open("C:/Users/HP/Desktop/Data_Sets/ETS2/Decision_tree_ten.txt",'w') as f:
    f=tree.export_graphviz(dtree,out_file=f,feature_names=cols)
```

### Accuracy score for DTC

```python
from sklearn.metrics import accuracy_score

x_train_pred=dtree.predict(x_train)
x_test_pred=dtree.predict(x_test)

print('Accuracy score for train data: ', accuracy_score(y_train,x_train_pred))
print('Accuracy score for test data: ', accuracy_score(y_test,x_test_pred))
```

### Confusion Matrix and Classification Report

```python
from sklearn.metrics import confusion_matrix, classification_report

# for train data
conf=confusion_matrix(y_train,x_train_pred)
print(conf)
cmap=sns.cubehelix_palette(50,hue=0.05,rot=0,light=0.9,dark=0,as_cmap=True)
sns.heatmap(conf,cmap=cmap,xticklabels=['0','1'],yticklabels=['0','1'],annot=True,square=True,fmt='d')
plt.show()
print('Classification Report: ', classification_report(y_train,x_train_pred))
```

```python
# for test data
conf=confusion_matrix(y_test,x_test_pred)
print(conf)
cmap=sns.cubehelix_palette(50,hue=0.05,rot=0,light=0.9,dark=0,as_cmap=True)
sns.heatmap(conf,cmap=cmap,xticklabels=['0','1'],yticklabels=['0','1'],annot=True,square=True,fmt='d')
plt.show()
print('Classification Report: ', classification_report(y_test,x_test_pred))
```

### Hyperparameter  of  DTC

#### criterion: {"gini","entropy","log-loss"}, default="gini"

#### splitter: {"best","random"}, default="best"

#### max_depth: int, default=None

#### min_samples_split: int or float, default=2

#### min_samples_leaf: int or float, int or float, default=1

#### min_weight_fraction_leaf: float, default=0.0

#### max_features: int, float or {"auto", "sqrt", "log2"}, default=None

#### random_state: int, RandomState Instance, None, default=None

#### max_leaf_nodes: int, default=None

#### ccp_alpha: non-negative float, default=0.0

###  Hyperparameter Tuning for DTC
