```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_excel('C:/Users/HP/Desktop/Data_Sets/ETS2/Transport_Data.xlsx')
```

```python
sns.distplot(data['Price Per KM'])
plt.title('Price per KM distribution')
plt.show()
```

```python
sns.distplot(data['Price'])
plt.title('Price distribution')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x = data['Transport Distance'], y = data['Price'])
plt.ylabel('Price', fontsize=13)
plt.xlabel('Distance', fontsize=13)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x = data['Cargo Weight'], y = data['Price'])
plt.ylabel('Price', fontsize=13)
plt.xlabel('Tons', fontsize=13)
plt.show()
```

```python
sns.relplot(x='Cargo Weight',y='Price',hue='Ferry?',data=data)
plt.show()
```
